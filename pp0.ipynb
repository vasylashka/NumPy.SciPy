```python
#NumPy
import numpy as np
matrix = np.random.uniform(3, 100, size=(15, 20))
print(matrix)
```

```python
def find_nearest_value(array, target_value):
    return matrix[np.unravel_index(np.argmin(np.abs(array - target_value)), array.shape)]
print("Nearest value to target: ", find_nearest_value(matrix,100))
```

```python
print("Average: ", np.mean(matrix))
print("Median: " , np.median(matrix))
```

```python
import matplotlib.pyplot as plt
plt.imshow(matrix)
plt.colorbar()
plt.show()
```

```python
plt.contourf(matrix)
plt.colorbar()
plt.show()
```

```python
def impute_missing_with_mean(array):
    column_means = np.nanmean(array,axis=0)
    nan_indices = np.isnan(array)
    matrix[nan_indices] = np.take(column_means, np.where(nan_indices)[1])
#filling with Nan   
x, y = np.random.randint(0, 15, size=(1, 10)), np.random.randint(0, 20, size=(1, 10))
for i,j in zip(x,y):
    matrix[i,j]=np.nan
#restoring values 
impute_missing_with_mean(matrix)
print(matrix)
```

```python
def normalize_matrix(array):
    min_val = np.min(array)
    max_val = np.max(array)
    return (array - min_val) / (max_val - min_val)
print(normalize_matrix(matrix))
```

```python
#SciPy part using only 
from PIL import Image

image = np.array(Image.open("sunflower.jfif"))
plt.imshow(image)
plt.title('Original image')
```

```python
def enhance_contrast(img):
    min_percentile = 2
    max_percentile = 98

    minvals = np.percentile(img, min_percentile, axis=(0, 1))
    maxvals = np.percentile(img, max_percentile, axis=(0, 1))

    adjusted_channels = []
    for channel in range(img.shape[2]):
        adjusted_channel = img[:, :, channel]
        adjusted_channel = np.clip(adjusted_channel, minvals[channel], maxvals[channel])
        adjusted_channel = ((adjusted_channel - minvals[channel]) / (maxvals[channel] - minvals[channel])) * 255
        adjusted_channels.append(adjusted_channel)

    adjusted = np.stack(adjusted_channels, axis=-1)

    return Image.fromarray(adjusted.astype(np.uint8))

adjusted_image = enhance_contrast(image)
plt.imshow(adjusted_image)
plt.title("Contrasted picture")
```

```python
grayscale_image = np.dot(image, [0.2989, 0.5870, 0.1140])
grayscale_image_pil = Image.fromarray(grayscale_image.astype(np.uint8))
plt.imshow(grayscale_image_pil, cmap='gray')
plt.title("Greyscale")
```

```python
import scipy.ndimage as ndimage

sigma = 20.0 
blurred_image = ndimage.gaussian_filter(image, sigma=(sigma, sigma, 0))
plt.imshow(blurred_image)
plt.title("Blurred")
```

```python
gradient = ndimage.morphological_gradient(grayscale_image, size=(3, 3))
plt.imshow(gradient, cmap='gray')
plt.title('Contour')
```
